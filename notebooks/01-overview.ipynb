## Quick Start

````python
from eliana.datasets import ParlogsObservations

config = {
    'system': 'PIONIER',
    'period': '1w'
}
logs = ParlogsObservations(**config)
````

## Meta data
The meta data referes to the high level task of activities with the list of observations in the selected period.

````python
meta = logs.meta
meta
````

## Traces
Traces are the event logs belonging to the specified observation in meta data table.

````python
# Contents of log trace with id=2 in logs.meta
trace = logs.load_trace(2)
trace
````

````python
print(f'logs  (type {type(logs)})')
print(f'meta  (type {type(meta)})')
print(f'trace (type {type(trace)})')
````

````python
# Estimate multiple loads of trace id=2
%timeit logs.load_trace(2)
````
